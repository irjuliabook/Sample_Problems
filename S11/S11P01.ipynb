```julia
using Plots
```

```julia
f1(x) = x;
f2(x) = sin(x);
f3(x) = x + sin(x);
```

```julia
Plots.plot(f1, 0:0.1:100, color="black", label="x", title="y = x", xlabel="x [m]", ylabel="y[m]")
```

```julia
Plots.plot(f2, 0:0.1:100, color = "black", label="sin(x)", title="y = sin(x)", xlabel="x [m]", ylabel="y[m]")
```

```julia
Plots.plot(f3, 0:0.1:100, color = "black", label="x + sin(x)", title="y = x + sin(x)", xlabel="x [m]", ylabel="y[m]")
```
